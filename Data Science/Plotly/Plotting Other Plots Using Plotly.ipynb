# Plotting Other Plots Using Plotly

```python
import numpy as np
```

```python
from matplotlib import pyplot
%matplotlib inline
```

```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
import pandas as pd
```

```python
cf.go_offline()
```

```python
init_notebook_mode(connected=True)
```

```python
df1=pd.DataFrame(np.random.randn(100,4),columns=['A','B','C','D'])
```

```python
df2=pd.DataFrame({'category':['A','B','C'],'values':[32,43,50]})
```

```python
df2.head()
```

```python
# Scatter Plot
```

```python
df1.iplot(kind="scatter",x='A',y='B',mode='markers')
```

```python
# Bar Plot ( For Bar plot x should be categorical, So Lets Plot df2)
```

```python
df2.iplot(kind="bar",x="category",y='values')
```

```python
# Our data is not always contain categorical column for those non categorical dataframe we can use aggregate functions
```

```python
# Here dataframe 1 is non categorical so lets plot df1
```

```python
df1.sum().iplot(kind='bar') # Using sum()
```

```python
df2.count().iplot(kind='box')
```

```python
# Box plot
```

```python
df1.iplot(kind='box')
```

```python
# Surface Plot - 3D 
```

```python
# lets create new dataframe
```

```python
df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[500,400,300,200,100]})
```

```python
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu') # colorscale=red yellow blue
```

```python
# Histogram
```

```python
df1['A'].iplot(kind='hist') #plotting the single column
```

```python
df1.iplot(kind='hist') #plotting all columns one over other
```

```python
#Spread
```

```python
df1[["A","B"]].iplot(kind="spread")
```

```python
# Bubble 
```

```python
df1.iplot(kind='bubble',x='A',y='B',size='C')
```
